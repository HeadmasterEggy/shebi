```python
# ===================== æ­£åˆ™è¡¨è¾¾å¼é¢„ç¼–è¯‘ =====================
TAG_RE = re.compile(r'<[^>]+>')
URL_RE = re.compile(r'http[s]?://(?:[a-zA-Z0-9\$\-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')
EMOJI_RE = re.compile(r'[\U0001F300-\U0001F9FF]|[\u2600-\u26FF]|[\u2700-\u27BF]')
PUNCT_RE = re.compile(r'[^\w\s\u4e00-\u9fffï¼Œã€‚ï¼ï¼Ÿã€]')
DIGITS_RE = re.compile(r'\d+')
REPEAT_CHARS_RE = re.compile(r'([\u4e00-\u9fff])\1{2,}')
SPACE_RE = re.compile(r'\s+')
# æ–°å¢ï¼šç”¨äºå»é™¤è‹±æ–‡å­—ç¬¦
ENGLISH_RE = re.compile(r'[a-zA-Z]+')


def clean_text(review):
    """
    å¯¹æ–‡æœ¬è¿›è¡Œæ¸…æ´—ï¼š
      - å»é™¤æ ‡ç­¾ã€é“¾æ¥ã€è¡¨æƒ…ã€æ ‡ç‚¹ã€æ•°å­—åŠé‡å¤å­—ç¬¦
      - å»é™¤è‹±æ–‡å­—ç¬¦
      - å»é™¤å¤šä½™ç©ºæ ¼
    """
    # å¤„ç† NaN å€¼
    if pd.isna(review):
        return ""

    # å°†éå­—ç¬¦ä¸²ç±»å‹è½¬æ¢ä¸ºå­—ç¬¦ä¸²
    if not isinstance(review, str):
        review = str(review)
    review = TAG_RE.sub('', review)
    review = URL_RE.sub('', review)
    review = EMOJI_RE.sub('', review)
    review = PUNCT_RE.sub('', review)
    review = DIGITS_RE.sub('', review)
    review = REPEAT_CHARS_RE.sub(r'\1', review)
    review = ENGLISH_RE.sub('', review)  # å»é™¤è‹±æ–‡
    review = SPACE_RE.sub(' ', review).strip()
    return review
```

```python
# æµ‹è¯•æ¸…æ´—å‡½æ•°
sample_text = "è¿™æ˜¯ä¸€ä¸ªæµ‹è¯•æ–‡æœ¬ï¼ŒåŒ…å«ä¸€äº›æ ‡ç‚¹ç¬¦å·ï¼123å’ŒURL https://example.com ä»¥åŠ<tag>æ ‡ç­¾</tag>ğŸ˜Š"

logging.info(f"\næµ‹è¯•æ¸…æ´—å‡½æ•°:")
logging.info(f"åŸæ–‡æœ¬: {sample_text}")
logging.info(f"æ¸…æ´—å: {clean_text(sample_text)}")
```

```python
# è¯»å–åœç”¨è¯
stopwords = []
with open("../stopword.txt", "r", encoding="utf-8") as f:
    for line in f.readlines():
        stopwords.append(line.strip())
```

```python
import jieba


def tokenize(review, stopwords):
    """
    å¯¹å•ä¸ªæ–‡æœ¬è¿›è¡Œåˆ†è¯ï¼Œå»é™¤åœç”¨è¯å’Œç©ºç™½å­—ç¬¦ï¼Œè¿”å›ä»¥ç©ºæ ¼åˆ†éš”çš„å­—ç¬¦ä¸²ã€‚
    """
    words = jieba.cut(review, cut_all=False)  # ä½¿ç”¨ç²¾ç¡®æ¨¡å¼è¿›è¡Œåˆ†è¯
    # è¿”å›ä¸€ä¸ªç©ºæ ¼åˆ†éš”çš„å­—ç¬¦ä¸²
    return ' '.join([word for word in words if word not in stopwords and word.strip()])
```

```python
# æµ‹è¯•åˆ†è¯å‡½æ•°
if sample_text:
    cleaned_sample = clean_text(sample_text)
    tokens = tokenize(cleaned_sample, stopwords)
    logging.info(f"åˆ†è¯ç»“æœ: {tokens}")
```

```python
from tqdm import tqdm


def process_texts(review, stopwords):
    """
    å¯¹æ‰€æœ‰æ–‡æœ¬è¿›è¡Œæ¸…æ´—å’Œåˆ†è¯ï¼Œå¹¶è¿”å›å¤„ç†åçš„è¯åˆ—è¡¨ã€‚
    """
    logging.info("å¼€å§‹æ¸…æ´—æ–‡æœ¬...")
    cleaned_texts = [clean_text(text) for text in tqdm(review, desc="æ¸…æ´—æ–‡æœ¬")]

    logging.info("å¼€å§‹åˆ†è¯å¤„ç†...")
    tokenized_texts = [tokenize(text, stopwords) for text in tqdm(cleaned_texts, desc="åˆ†è¯å¤„ç†")]

    return tokenized_texts
```

```python
processed_texts = process_texts(df.review, stopwords)
processed_texts
```

```python
combined_data = list(zip(df['label'], processed_texts))
```

```python
with open('../trained.txt', 'w', encoding='utf-8') as file:
    for label, text in combined_data:
        file.write(f"{label} {text}\n")
```
